```python
#  LE VISUALIZZAZIONI SEARBON basato su libreria matplotlib
import matplotlib.pyplot as plt   
import seaborn as sns
plt.rcParams['figure.figsize'] = 8,4 # parametri editabili online trovi docs
```

```python
stats.head()
```

```python
visl = sns.displot(stats['InternetUsers'], bins=5 )# primo par var
```

```python
vis2 = sns.boxplot(data = stats, x='IncomeGroup', y='Birthrate')  #rapporto(plot) nascte e gruppi reddito, reddito alto
                                                                    # nascite basse
```

```python
# creiamo un nuovo plot arricchiamolo di parametri
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='Birthrate', fit_reg = False, hue='IncomeGroup', size=5, aspect=2)  
#. shift tab per vedere info funzione
#fig_reg toglie linea
# hue color
# size controlla la dimensione del grafico
# aspect con 5 prende l'altezza
# dimensione marker con le keywords arguments
# quando non c'è internet si fanno + figli non dipende solo da quello ma incide a quanto pare
```

```python
#Keywords Arguments   ARGOMENTO AVANZATO
# marker size ha bisogno di un parametro
# dizionari scatter_kws e line_kws parametri addizionali da poter passare a plt.scatter e plt.plot 
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='Birthrate', fit_reg = False, hue='IncomeGroup', size=5, \
                   aspect=2 ,scatter_kws={'s':200})# passiamo il dizionario che è un elemento 
                                        # chiave s valore 100 inoltrandolo a lmplot  
 
```
